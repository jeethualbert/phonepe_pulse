```python
# IMPORT LIBRARIES

import os
import json
import mysql.connector
import pandas as pd
```

```python
# HELPER FUNCTION TO CLEAN STATE NAMES

def clean_state_names(df):

    df["States"] = df["States"].str.replace("andaman-&-nicobar-islands", "Andaman & Nicobar")
    df["States"] = df["States"].str.replace("-", " ")
    df["States"] = df["States"].str.title()
    df["States"] = df["States"].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli and Daman and Diu")

    return df
```

```python
# AGGREGATED INSURANCE DATA EXTRACTION
# ------------------------------------

# aggregated_insurance_path = "pulse/data/aggregated/insurance/country/india/state/"
aggregated_insurance_path = "C:/vscode/phonepe/pulse/data/aggregated/insurance/country/india/state"

insurance_data = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "Insurance_type": [],
    "Insurance_count": [],
    "Insurance_amount": []
}

for state in os.listdir(aggregated_insurance_path):
    state_path = os.path.join(aggregated_insurance_path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            file_path = os.path.join(year_path, file)

            with open(file_path, "r") as json_file:
                json_data = json.load(json_file)
                if json_data["data"]["transactionData"] is not None:
                    for item in json_data["data"]["transactionData"]:
                        insurance_data["States"].append(state)
                        insurance_data["Years"].append(int(year))
                        insurance_data["Quarter"].append(int(file.strip(".json")))
                        insurance_data["Insurance_type"].append(item["name"])
                        insurance_data["Insurance_count"].append(item["paymentInstruments"][0]["count"])
                        insurance_data["Insurance_amount"].append(item["paymentInstruments"][0]["amount"])

# Convert to dataframe
aggregated_insurance_df = pd.DataFrame(insurance_data)

# Clean state names
aggregated_insurance_df = clean_state_names(aggregated_insurance_df)

aggregated_insurance_df.head()
```

```python
# AGGREGATED TRANSACTION DATA EXTRACTION
# ----------------------------------------

aggregated_transaction_path = "C:/vscode/phonepe/pulse/data/aggregated/transaction/country/india/state"

transaction_data = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "Transaction_type": [],
    "Transaction_count": [],
    "Transaction_amount": []
}

for state in os.listdir(aggregated_transaction_path):
    state_path = os.path.join(aggregated_transaction_path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            file_path = os.path.join(year_path, file)

            with open(file_path, "r") as data_file:
                json_data = json.load(data_file)

                if json_data["data"]["transactionData"] is not None:

                    for transaction in json_data["data"]["transactionData"]:
                        transaction_data["States"].append(state)
                        transaction_data["Years"].append(int(year))
                        transaction_data["Quarter"].append(int(file.strip(".json")))
                        transaction_data["Transaction_type"].append(transaction["name"])
                        transaction_data["Transaction_count"].append(transaction["paymentInstruments"][0]["count"])
                        transaction_data["Transaction_amount"].append(transaction["paymentInstruments"][0]["amount"])


aggregated_transaction_df = pd.DataFrame(transaction_data)
aggregated_transaction_df = clean_state_names(aggregated_transaction_df)
aggregated_transaction_df.head()
```

```python
# AGGREGATED USER DATA EXTRACTION
# -----------------------------------

aggregated_user_path = "C:/vscode/phonepe/pulse/data/aggregated/user/country/india/state"

user_data = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "Brands": [],
    "Transaction_count": [],
    "Percentage": []
}

for state in os.listdir(aggregated_user_path):
    state_path = os.path.join(aggregated_user_path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            file_path = os.path.join(year_path, file)

            with open(file_path, "r") as json_file:
                json_data = json.load(json_file)

                try:
                    for item in json_data["data"]["usersByDevice"]:
                        user_data["States"].append(state)
                        user_data["Years"].append(int(year))
                        user_data["Quarter"].append(int(file.strip(".json")))
                        user_data["Brands"].append(item["brand"])
                        user_data["Transaction_count"].append(item["count"])
                        user_data["Percentage"].append(item["percentage"])
                except:
                    pass


aggregated_user_df = pd.DataFrame(user_data)
aggregated_user_df = clean_state_names(aggregated_user_df)

aggregated_user_df.head()
```

```python
# MAP INSURANCE DATA EXTRACTION
# ------------------------------

map_insurance_path = "C:/vscode/phonepe/pulse/data/map/insurance/hover/country/india/state"

map_insurance_data = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "District": [],
    "Transaction_count": [],
    "Transaction_amount": []
}


# Loop through states
for state in os.listdir(map_insurance_path):
    state_path = os.path.join(map_insurance_path, state)

    # Loop through years
    for year in os.listdir(state_path):

        year_path = os.path.join(state_path, year)

        # Loop through quarter JSON files
        for file in os.listdir(year_path):

            file_path = os.path.join(year_path, file)

            with open(file_path, "r") as json_file:
                json_data = json.load(json_file)

                # Extract district level insurance data
                for item in json_data["data"]["hoverDataList"]:

                    map_insurance_data["States"].append(state)
                    map_insurance_data["Years"].append(int(year))
                    map_insurance_data["Quarter"].append(int(file.strip(".json")))
                    map_insurance_data["District"].append(item["name"])
                    map_insurance_data["Transaction_count"].append(item["metric"][0]["count"])
                    map_insurance_data["Transaction_amount"].append(item["metric"][0]["amount"])


# Convert to dataframe
map_insurance_df = pd.DataFrame(map_insurance_data)

# Clean state names
map_insurance_df = clean_state_names(map_insurance_df)

map_insurance_df.head()
```

```python
# MAP USER DATA EXTRACTION
# --------------------------

map_user_path = "C:/vscode/phonepe/pulse/data/map/user/hover/country/india/state"

map_user_data = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "District": [],
    "RegisteredUser": [],
    "AppOpens": []
}

for state in os.listdir(map_user_path):
    state_path = os.path.join(map_user_path, state)

    for year in os.listdir(state_path):
        year_path = os.path.join(state_path, year)

        for file in os.listdir(year_path):
            file_path = os.path.join(year_path, file)

            with open(file_path, "r") as json_file:
                json_data = json.load(json_file)

                # hoverData is dictionary
                for district, data in json_data["data"]["hoverData"].items():
                    map_user_data["States"].append(state)
                    map_user_data["Years"].append(int(year))
                    map_user_data["Quarter"].append(int(file.strip(".json")))
                    map_user_data["District"].append(district)
                    map_user_data["RegisteredUser"].append(data["registeredUsers"])
                    map_user_data["AppOpens"].append(data["appOpens"])

map_user_df = pd.DataFrame(map_user_data)
map_user_df = clean_state_names(map_user_df)

map_user_df.head()
```

```python
#  CONNECT TO MYSQL DATABASE
```

```python
import mysql.connector

mysql_connection = mysql.connector.connect(
    host="localhost",
    user="root",
    password="root"
)

mysql_cursor = mysql_connection.cursor()

# Create and use database
mysql_cursor.execute("CREATE DATABASE phonepe_data")
mysql_cursor.execute("USE phonepe_data")
```

```python
# CREATE TABLE: AGGREGATED INSURANCE
# -------------------------------------

create_aggregated_insurance_table = """
CREATE TABLE IF NOT EXISTS aggregated_insurance (
    state_name VARCHAR(50),
    year INT,
    quarter INT,
    insurance_type VARCHAR(50),
    insurance_count BIGINT,
    insurance_amount BIGINT
)
"""

mysql_cursor.execute(create_aggregated_insurance_table)


# Insert aggregated insurance data
insert_aggregated_insurance = """
INSERT INTO aggregated_insurance
(state_name,year,quarter,insurance_type,insurance_count,insurance_amount)
VALUES (%s,%s,%s,%s,%s,%s)
"""

for index, row in aggregated_insurance_df.iterrows():

    values = (
        row["States"],
        row["Years"],
        row["Quarter"],
        row["Insurance_type"],
        row["Insurance_count"],
        row["Insurance_amount"]
    )

    mysql_cursor.execute(insert_aggregated_insurance, values)

mysql_connection.commit()
```

```python
# CREATE TABLE: AGGREGATED TRANSACTION
# ----------------------------------------

create_aggregated_transaction_table = """
CREATE TABLE IF NOT EXISTS aggregated_transaction (
    state_name VARCHAR(50),
    year INT,
    quarter INT,
    transaction_type VARCHAR(50),
    transaction_count BIGINT,
    transaction_amount BIGINT
)
"""

mysql_cursor.execute(create_aggregated_transaction_table)


insert_aggregated_transaction = """
INSERT INTO aggregated_transaction
(state_name,year,quarter,transaction_type,transaction_count,transaction_amount)
VALUES (%s,%s,%s,%s,%s,%s)
"""

for index, row in aggregated_transaction_df.iterrows():

    values = (
        row["States"],
        row["Years"],
        row["Quarter"],
        row["Transaction_type"],
        row["Transaction_count"],
        row["Transaction_amount"]
    )

    mysql_cursor.execute(insert_aggregated_transaction, values)

mysql_connection.commit()
```
